```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

### Visualize Microsoft and Apple stock price from 2006-01-01 to 2018-01-01

```python
os.getcwd()
```

Check the data

```python
os.chdir("C:\\Users\\shang\\Desktop\\Python\\stock-time-series")
msft = pd.DataFrame(pd.read_csv("MSFT_2006-01-01_to_2018-01-01.csv"))
msft['Date'] = pd.to_datetime(msft['Date'])
msft
```

Get the names of varibles

```python
list(msft)
```

1. Line plots of different prices from 2006 to 2018. They appear to overlap.

```python
oM = msft['Open']
hM = msft['High']
lM = msft['Low']
cM = msft['Close']
```

```python
fig, ax = plt.subplots(1, 4, figsize=(17,4))
ax[0].plot(oM)
ax[0].set_title("Open Price")
ax[1].plot(hM, color='green')
ax[1].set_title("High Price")
ax[2].plot(lM, color='red')
ax[2].set_title("Low Price")
ax[3].plot(cM, color='black')
ax[3].set_title("Close Price")
```

Get the Average price of the Open, High, Low and Close (OHLC Average)

```python
msft['avgM'] = (oM+hM+lM+cM)/4
```

2.The distplot of the average price

```python
a = msft['avgM']
fig, ax = plt.subplots(figsize=(11,8))
sns.distplot(a, kde_kws={'color': 'b'}, axlabel = 'Average price')
```

3.Extract year from Date. Boxplot of Average price by years

```python
msft['year'] = msft['Date'].dt.year
fig, ax = plt.subplots(figsize=(11,8))
sns.set_style('whitegrid')
ax = sns.boxplot(x='year', y='avgM', data=msft)
ax.set_xlabel('Year',size=15)
ax.set_ylabel('Average Price',size=15)
```

Import Apple Stock data

```python
aapl = pd.DataFrame(pd.read_csv('AAPL_2006-01-01_to_2018-01-01.csv'))
aapl['Date'] = pd.to_datetime(aapl['Date'])
oA = aapl['Open']
hA = aapl['High']
lA = aapl['Low']
cA = aapl['Close']
aapl['avgA'] = (oA+hA+lA+cA)/4
aapl['year'] = aapl['Date'].dt.year
aapl.head(10)
```

4.Violin Plot -- Compare Microsoft with Apple

```python
df = pd.concat([msft.melt(id_vars='year', value_vars='avgM'),
                aapl.melt(id_vars='year', value_vars='avgA')], ignore_index=True)
plt.subplots(figsize=(15,8))
ax = sns.violinplot(x='year', y='value', data=df, hue='variable',split=True, palette='Set1')
ax.set_xlabel('Year', size=15)
ax.set_ylabel('Average Price', size=15)
ax.set_title('Compare Average Price of Microsoft and Apple', size=20)
l = ax.legend(loc='upper left', prop={'size':16})
l.get_texts()[0].set_text('Microsoft')
l.get_texts()[1].set_text('Apple')
```

5.Compare stock Average Price of two companys.

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot('avgM', data=msft)
ax.plot('avgA', data=aapl)
ax.set_title('Stock Average Price from 2006 to 2018',fontsize=16)
l = ax.legend(prop={'size':15})
l.get_texts()[0].set_text('Microsoft')
l.get_texts()[1].set_text('Apple')
```
